# miremos una taba de todos los tipos para ver cual es el ataque mas alto

# con esto podemos tener un equipo con los pokemón mas fuertes, otros con  mas defensa, y por ultimo unos con mas velocidad, pero no esta tan equilibrado, que tal si ahora sumamos todas las estadisticas y buscamos los mejores

```python
pk['Total'] = pk.iloc[:, 4:10].sum(axis=1)
pk
```

```python
total=pk.sort_values(['Total'], ascending=False)[0:10]
total= total.reset_index()
total
```

```python
width = 0.25
plt.bar(total['Name'], total['Total'], label='PKMN Total', width=width, color='blue')
plt.bar(total['Name'], total['Attack'], label='PKMN Atk', width=width, color='orange')
plt.bar(total['Name'], total['Speed'], label='PKMN Sd', width=width, color='green')
plt.bar(total['Name'], total['HP'], label='PKMN HP', width=width, color='red')
plt.title("PKMN")
plt.legend()
plt.xlabel("Names")
plt.ylabel("Stats")
plt.xticks(rotation=90)
```

# bien los tenemos, estos pueden cosntituir el equipo mas poderoso, pero todos son lejendarios, eso hace que el equipo sea muy dificl de conseguir, que tal el equipo mas poderoso de pokemón que no son legendarios?

```python
nole=pk.loc[pk['Legendary'] == False].sort_values(['Total'], ascending=False)[0:10]
nole= nole.reset_index()
nole
```

```python
width = 0.25
plt.bar(nole['Name'], nole['Total'], label='PKMN Total', width=width, color='blue')
plt.bar(nole['Name'], nole['Attack'], label='PKMN Atk', width=width, color='orange')
plt.bar(nole['Name'], nole['Speed'], label='PKMN Sd', width=width, color='green')
plt.bar(nole['Name'], nole['HP'], label='PKMN HP', width=width, color='red')
plt.title("PKNL")
plt.legend()
plt.xlabel("Names")
plt.ylabel("Stats")
plt.xticks(rotation=90)
```
